```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import astropy.units as u
from astropy.constants import L_sun, sigma_sb, R_sun
```

```python
Tsublim = 1500 * u.K
```

```python
F_sun = sigma_sb * (5000*u.K)**4
```

```python
F_sun.to("W/m^2") 
```

```python
def F(a):
    return F_sun
```

```python
Tstar = np.linspace(3000, 5000, 100) * u.K
R = np.linspace(0.1, .9, 100) * u.R_sun
a = np.linspace(.005,.15,100) * u.AU
```

```python
aeq = []
for (t, r) in zip(Tstar,R):
    res = (t*np.sqrt(r/a)).decompose()
    plt.plot(a, res)
    aeq.append(a[np.argmin((res-Tsublim)**2)])
plt.ylim(0,3000)
plt.xlabel("a [AU]")
plt.ylabel("$T_\mathrm{eq}$ [K]")
```
