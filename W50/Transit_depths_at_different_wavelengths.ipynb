```python
import numpy as np
from astropy.modeling import blackbody as bb 
from astropy.constants import R_sun, h, c, k_B
import astropy.units as u

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lambds = (400*u.nm, 1000*u.nm, 3500*u.nm)
frac_spot = .3
for lambd in lambds:
    Fstar = bb.blackbody_lambda(lambd, 5000*u.K)
    Fspot = bb.blackbody_lambda(lambd, 3000*u.K)
    F_out = Fstar * (1 - frac_spot) + Fspot * frac_spot
    F_in = Fstar * (1 - frac_spot - 0.025) + Fspot * frac_spot 
    print(lambd, (F_out-F_in)/F_in*100)
```

```python
l = np.linspace(300,3300,1000)*u.nm
plt.plot(l,bb.blackbody_lambda(l, 5000*u.K) )
plt.plot(l,bb.blackbody_lambda(l, 3000*u.K) )
plt.yscale("log")
```
